# Filling Area on Line Plots

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/05-Fill_Betweens/data.csv'
data = pd.read_csv(url)
data.head()
```

```python
plt.style.use("fivethirtyeight")
ages = data['Age']
All_devs_salary = data['All_Devs']
python_devs_salary = data['Python']
javascript_devs_salary = data['JavaScript']

plt.title("Different Devs Salary By their Ages")
plt.xlabel("Ages")
plt.ylabel("Salary")

plt.plot(ages,All_devs_salary,color='b',label='All_Devs_Salary')
plt.plot(ages,python_devs_salary,color='g',linestyle='--',label='Python_Devs_Salary')
plt.plot(ages,javascript_devs_salary,color='y',linestyle=':',label='JavaScript_Devs_Salary')

plt.tight_layout()
plt.legend(title="Avaliable_Devs")
plt.show()
```

```python
#fill the plot
ages = data['Age']
All_devs_salary = data['All_Devs']
python_devs_salary = data['Python']
javascript_devs_salary = data['JavaScript']

plt.title("Different Devs Salary By their Ages")
plt.xlabel("Ages")
plt.ylabel("Salary")

plt.plot(ages,All_devs_salary,color='b',label='All_Devs_Salary')
plt.plot(ages,python_devs_salary,color='g',linestyle='--',label='Python_Devs_Salary')

'''
overal_median = 57287
plt.fill_between(ages,python_devs_salary,overal_median,where=(python_devs_salary>overal_median),interpolate=True,alpha=0.25,)
'''

plt.fill_between(ages,python_devs_salary,All_devs_salary,where=(python_devs_salary>All_devs_salary),interpolate=True,alpha=0.25,label='Above Avg')
plt.fill_between(ages,python_devs_salary,All_devs_salary,where=(python_devs_salary<=All_devs_salary),interpolate=True,alpha=0.25,label='Below Avg')


plt.tight_layout()
plt.legend()
plt.show()
```
