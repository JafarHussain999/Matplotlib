```python
import matplotlib
print(matplotlib.__version__) # To check the installed vesion
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints,linestyle='--',color='b',marker='1') # linestyle and color or marker the order can be anything 
plt.title("Basic Ploting")
plt.show()
```
