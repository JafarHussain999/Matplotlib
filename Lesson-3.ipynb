# Pie Charts

```python
from matplotlib import pyplot as plt
```

```python
plt.style.use("fivethirtyeight")
#basic pie chart
slice=[60,20,20]
label=["Windows","Linux","Mac"]
colors=['Red','Black','Blue'] #use the hexa value of color to look better
plt.pie(slice,labels=label,colors=colors,wedgeprops={'edgecolor':'black'})
plt.title("Pie chart of OS")
plt.tight_layout()
plt.show()
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0] #one slice get out of the plot like in python

plt.pie(slices, labels=labels, explode=explode, shadow=True,startangle=90, autopct='%1.1f%%',wedgeprops={'edgecolor': 'black'})

#autopct is used to print the percentage in the plot
#shadow is used to get shadow in the plot and look like the 3D
plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```
