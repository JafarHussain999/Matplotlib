# Scatter Plots

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# print(plt.style.available) #this are style avaliable in the matplotlib
plt.style.use('seaborn')
```

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
          538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]
plt.scatter(x,y,s=sizes,c=colors,cmap='Greens',marker='o',edgecolors='b',linewidths=1,alpha=0.7) #here S is size and c is color


cbar = plt.colorbar() #it is used side bar of colors
cbar.set_label("Satisfaction")

plt.title("Basic Scatter Plot")

plt.tight_layout()
plt.show()
```

```python
#Real time data 
url = 'https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/07-ScatterPlots/2019-05-31-data.csv'
data = pd.read_csv(url)
data.head()
```

```python
view_count = data['view_count']
likes      = data['likes']
ratio      = data['ratio']

plt.scatter(view_count,likes,c=ratio,cmap='Greens',edgecolors='k',linewidths=1)

cbar = plt.colorbar()
cbar.set_label('like and dislike Ratio')

plt.xlabel('log')
plt.ylabel('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()
plt.show()
```
